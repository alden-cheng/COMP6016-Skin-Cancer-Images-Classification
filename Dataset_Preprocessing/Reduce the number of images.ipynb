```python
import os
import shutil
import random
```

# Split the original dataset in half and put it into a new dataset

```python
src_folder = "/content/drive/MyDrive/Datasets/Ham10000/ham10000_images_classified"
output_folder = "/content/drive/MyDrive/Datasets/Ham10000/ham10000_reduction"

keep_ratio = 0.2

categories = ['nv', 'mel', 'bkl', 'bcc', 'akiec', 'vasc', 'df']

for category in categories:
    category_src_folder = os.path.join(src_folder, category)
    category_output_folder = os.path.join(output_folder, category)

    all_files = [f for f in os.listdir(category_src_folder) if os.path.isfile(os.path.join(category_src_folder, f))]

    selected_files = random.sample(all_files, int(len(all_files) * keep_ratio))

    os.makedirs(category_output_folder, exist_ok=True)

    for file in selected_files:
        shutil.copy(os.path.join(category_src_folder, file), os.path.join(category_output_folder, file))
```
